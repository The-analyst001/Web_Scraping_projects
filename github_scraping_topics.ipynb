# Scraping Popular Github Topics

Scrape the webpage https://github.com/topics and get the topics titles, its description, and the link towards it directs the users 

```python
# get the url webpage 
url = 'https://github.com/topics'

# import the necessary library and read the webpage as a html file 
from bs4 import BeautifulSoup
import requests

page = requests.get(url)

webpage = BeautifulSoup(page.text, 'html')

# print the first 1000 characters of the webpage, to have an overview...
print(page.text[:1000])
```

```python
# find and print each topic titles from the webpage in a list called topic_titles_list

topic_titles = webpage.find_all('p', class_= "f3 lh-condensed mb-0 mt-1 Link--primary")

topic_titles_list = [item.text for item in topic_titles]

print(topic_titles_list)
```

```python
# find extract the topic descriptions 

topic_descriptions = webpage.find_all('p', class_= "f5 color-fg-muted mb-0 mt-1")

topic_descriptions_list = [item.text.strip() for item in topic_descriptions]

print(topic_descriptions_list)
```

```python
# find extract the topic url

topic_url = webpage.find_all('a', class_= "no-underline flex-1 d-flex flex-column")

index = len(topic_url)
#print(topic_url[0]['href'])

topic_url_list = [ url+topic_url[i]['href'] for i in range(index)]

print(topic_url_list)

```

```python
# create a dataframe containing the extracted elements
dictionnary = { 'Topic_titles': topic_titles_list,
              'Topic_descriptions': topic_descriptions_list,
              'Topic_url': topic_url_list }

import pandas as pd
df = pd.DataFrame(dictionnary)
display(df)
```
